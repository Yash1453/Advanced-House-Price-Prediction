```python
from warnings import filterwarnings
filterwarnings("ignore")

import pandas as pd
A = pd.read_csv("C:/PythonFiles/Cars93.csv")
```

```python
from preprocessing_6PM import replacer
replacer(A)
```

```python
Y = A[["Weight"]]
X = A.drop(labels=["id","Make","Model","Manufacturer","Weight"],axis=1)

from sklearn.preprocessing import StandardScaler
cat = []
con = []
for i in X.columns:
    if(A[i].dtypes=="object"):
        cat.append(i)
    else:
        con.append(i)
        
X1 = pd.get_dummies(X[cat])
ss = StandardScaler()
X2 = pd.DataFrame(ss.fit_transform(X[con]),columns=con)
Xnew = X2.join(X1)
from statsmodels.api import OLS,add_constant
Xconst = add_constant(Xnew)
ols = OLS(Y,Xconst)
model = ols.fit()
model.summary()
```

```python
Q = pd.DataFrame(model.pvalues,columns=["pvals"])
```

```python
x = Q.pvals.max()
col_to_drop = Q[Q.pvals == x].index[0]
print(col_to_drop)
```

```python
x = Q.pvals.max()
col_to_drop = Q[Q.pvals == x].index[0]
print(col_to_drop)

Xnew = Xnew.drop(labels=[col_to_drop],axis=1)

from statsmodels.api import OLS,add_constant
Xconst = add_constant(Xnew)
ols = OLS(Y,Xconst)
model = ols.fit()
model.summary()
```

```python
Q = pd.DataFrame(model.pvalues,columns=["pvals"])
x = Q.pvals.max()
col_to_drop = Q[Q.pvals == x].index[0]
print(col_to_drop)

Xnew = Xnew.drop(labels=[col_to_drop],axis=1)

from statsmodels.api import OLS,add_constant
Xconst = add_constant(Xnew)
ols = OLS(Y,Xconst)
model = ols.fit()
model.summary()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=21)

from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model = lm.fit(xtrain,ytrain)

pred_tr=model.predict(xtrain)
pred_ts=model.predict(xtest)

from sklearn.metrics import mean_squared_error
tr_err = mean_squared_error(ytrain,pred_tr)
ts_err = mean_squared_error(ytest,pred_ts)
```

# Regularized Ridge Model

```python
from sklearn.linear_model import Ridge
rr = Ridge()

tg = {'alpha':[0.999,0.998,0.997,0.996,0.995,0.994]}

from sklearn.model_selection import GridSearchCV
cv = GridSearchCV(rr,tg,scoring="neg_mean_squared_error",cv=4)

cvmodel = cv.fit(xtrain,ytrain)
cvmodel.best_params_
```

```python
c = cvmodel.best_params_['alpha']
```

```python
rr = Ridge(alpha=c)
model = rr.fit(xtrain,ytrain)
pred_tr=model.predict(xtrain)
pred_ts=model.predict(xtest)

from sklearn.metrics import mean_squared_error
tr_err = mean_squared_error(ytrain,pred_tr)
ts_err = mean_squared_error(ytest,pred_ts)
```
